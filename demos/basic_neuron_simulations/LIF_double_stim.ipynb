```python
import os, sys
sys.path.append('../../')

import numpy as np
import matplotlib.pyplot as plt

import neurons
```

```python
net = neurons.Network()

net.add(neurons.Sensory, 5, 'input_1')
net.add(neurons.Sensory, 5, 'input_2')
net.add(neurons.LIF, 5, 'neurons',
    v_rest=-70,
    v_thres=-55,
    v_reset=-70,
    R_r = 20
)
net.connect('input_1', 'neurons')
net.connect('input_2', 'neurons')
net.compile()
```

```python
net.train = True
```

```python
seconds = 10
dt = 0.001 # 1 ms


net.reset_history()
steps = int(seconds / dt)

for i in range(steps):
    if i % 20 == 0:
        if 1 <= i*dt < 4:
            net.bundles['input_1'].outputs[:] = 1
            net.bundles['input_2'].outputs[:] = 0
        if 4 <= i*dt < 5:
            net.bundles['input_1'].outputs[:] = 0
            net.bundles['input_2'].outputs[:] = 0
        if 5 <= i*dt < 8:
            net.bundles['input_1'].outputs[:] = 0
            net.bundles['input_2'].outputs[:] = 1
        if 8 <= i*dt < 10:
            net.bundles['input_1'].outputs[:] = 0
            net.bundles['input_2'].outputs[:] = 0
    else:
        net.bundles['input_1'].outputs[:] = 0.0
        net.bundles['input_2'].outputs[:] = 0.0
        
    net.step(dt)
```

```python
plt.figure(figsize=(14,5), dpi=80)
plt.title('input current')
plt.plot(np.arange(steps) * dt, net.history['neurons'].current)
plt.ylabel('current [mA]')
plt.xlabel('time [s]')
plt.show()
```

```python
plt.figure(figsize=(14,5), dpi=80)
plt.title('membrane potential')
plt.plot(np.arange(steps) * dt, net.history['neurons'].voltage)
plt.ylabel('voltage [mV]')
plt.xlabel('time [s]')
plt.ylim(-90,-50)
# plt.xlim(5,6)
plt.show()
```

```python
def get_rates(data, width, dt):
    rates = []
    kernel = np.kaiser(width, 50)
    kernel /= kernel.sum() * dt
    for x in data.T:
        rates.append(np.convolve(x, kernel, 'same'))
    return np.stack(rates).T
```

```python
plt.figure(figsize=(14,5), dpi=80)
plt.title('Spikes')
plt.plot(np.arange(steps) * dt, get_rates(net.history['neurons'].outputs, 1000, dt))
plt.ylabel('firing rate [Hz]')
plt.xlabel('time [s]')
# plt.ylim(0,1)
plt.show()
```
